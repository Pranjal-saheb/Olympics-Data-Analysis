```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load datasets
athletes = pd.read_csv("athlete_events.csv")
regions = pd.read_csv("noc_regions.csv")
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#join the dataframes using pandas merge functions
athletes_df = athletes.merge(regions,how = "left",on = "NOC")   ### noc is common column
athletes_df
```

```python
athletes_df.shape            #shape attribute
```

```python
athletes_df.rename(columns = {"region":"Region","notes":"Notes"},inplace = True)
athletes_df.head()
```

```python
athletes_df.info()
```

```python
#statistical parameterso
athletes_df.describe()
```

```python
nan_values = athletes_df.isna()
nan_columns = nan_values.any()
nan_columns
```

```python
# Top 10 countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
```

```python
#plot for the top 10 countries
plt.figure(figsize = (12,6))
plt.title("Overall participation by country")
sns.barplot(x = top_10_countries.index,y = top_10_countries,palette = "Set2")
```

```python
#Age distribution of athletes
plt.figure(figsize = (12,6))
plt.title("Age distribution of athletes")
plt.xlabel("Number of participants")
plt.ylabel("Age")
plt.hist(athletes_df.Age,bins = np.arange(10,80,2),color = "orange",edgecolor = "white");
```

```python
## winter season sports
winter_sports = athletes_df[athletes_df.Season == "Winter"].Sport.unique()
winter_sports
```

```python
## summer season sports
summer_sports = athletes_df[athletes_df.Season == "Summer"].Sport.unique()
summer_sports
```

```python
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
#Pie plot for male and female atheletes
plt.figure(figsize = (12,6))
plt.title("Gender distribution")
plt.pie(gender_counts,labels = gender_counts.index,startangle = 150,autopct = "%1.1f%%")
plt.show()
```

```python
# total number of female athletes in summer season
female_participants = athletes_df[(athletes_df.Sex == "F") & (athletes_df.Season == "Summer")][["Sex","Year"]]
female_participants = female_participants.groupby("Year").count().reset_index()
female_participants.head()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == "F") & (athletes_df.Season == "Summer")]
```

```python
sns.set(style = "darkgrid")
plt.figure(figsize = (20,10))
sns.countplot(x="Year",data = womenOlympics, palette = "Spectral")
plt.title("Women Participation")
```

```python
part = womenOlympics.groupby("Year")["Sex"].value_counts()
plt.figure(figsize = (20,10))
part.loc[:,"F"].plot()
plt.title("Plot of female athletes over time")
```

```python
# gold medal athletes
gold_medals = athletes_df[athletes_df.Medal == "Gold"]
gold_medals.head()
```

```python
# take only that values that are different from NaN
gold_medals = gold_medals[np.isfinite(gold_medals["Age"])]
```

```python
#gold medals beyond age 60
gold_medals["ID"][gold_medals["Age"] > 60].count()
```

```python
# sports in which gold medal taken by athlete whose age is beyond 60
sporting_event = gold_medals["Sport"][gold_medals["Age"] > 60]
sporting_event
```

```python
# plot for sporting event
plt.figure(figsize = (10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold medals for athletes over 60 years")
```

```python
# gold medals from each country
gold_medals.Region.value_counts().reset_index(name="Medal").head()
```

```python
totalGold_medals = gold_medals.Region.value_counts().reset_index(name="Medal").head(6)
g = sns.catplot(x="index", y ="Medal", data = totalGold_medals, height = 5,kind = "bar",palette = "rocket")
g.despine(left = True)
g.set_xlabels("Top 6 countries")
g.set_ylabels("Number of Medals")
plt.title("gold medals per country")
```

```python
## 2016 Rio Summer Olympics
max_year = athletes_df.Year.max()
print(max_year)
team_names = athletes_df[(athletes_df.Year == "max_year") & (athletes_df.Medal == "Gold")].Team
team_names.value_counts().head(10)
```

```python
## scatter plot of height vs weight of medalist
not_null_medals = athletes_df[(athletes_df["Height"].notnull()) & (athletes_df["Weight"].notnull())]
plt.figure(figsize = (20,10))
axis = sns.scatterplot(x="Height",y = "Weight",data = not_null_medals,hue = "Sex")
plt.title("Height vs Weight of Olympic Medalist")
```
